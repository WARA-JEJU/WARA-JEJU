```python
new_part = pd.concat([part2, part3])
```

```python
print(new_part.shape)
print(list(new_part.columns))
new_part.head()
```

```python
# 누락 데이터 확인

new_part.count()
```

```python
# 누락 데이터 확인

new_part.isnull().sum()
```

```python
# '구분'을 제외한 중복데이터 여부를 파악하기 위해 진행

a = new_part.iloc[:,[1,2,3,4,5,6]]
a
```

```python
# 완전히 동일한 중복데이터 여부 확인

a.duplicated().sum()
```

```python
# '데이터기준일자', '비고'를 제외한 중복 데이터 삭제
# '연락처' 항목은 결측값이 여럿 존재해서 제외

a.drop_duplicates(['상호명', '도로명주소', '객실수', '위도', '경도'])
```

```python
# 중복 데이터 개수 파악

a[a.duplicated()==True]
```
