```python
import pandas as pd
import numpy as np
  
part1 = pd.read_csv('/content/drive/MyDrive/국민대학원/졸업프로젝트/데이터/1순위/음식/대규모점포정보.csv', encoding='cp949')
```

# part1 = 대규모점포정보

```python
# part1 = 대규모점포정보

print(part1.shape)
print(list(part1.columns))
part1.head()
```

```python
# 컬럼 삭제

part1 = part1.drop(['업종구분대분류', '인허가일자', '인허가취소일자', '폐업일자', 
                    '휴업시작일자', '휴업종료일자', '재개업일자', '소재지면적', '지번주소', '도로명우편번호', '데이터갱신일자'], axis=1)
part1.head()
```

```python
# 컬럼명 변경

part1.rename(columns = {'업종구분소분류':'업종',},inplace=True)
part1.head()
```

```python
# 컬럼별로 중복데이터 여부 확인

part1['영업상태명'].value_counts()
```

```python
# '영업상태명' 컬럼에 있는 '폐업'이 들어가 행을 삭제

part1 = part1[part1.영업상태명 != '폐업']
part1.head()
```

```python
# '영업상태명' 컬럼에 있는 '폐업'이 들어가 행을 삭제

part1 = part1[part1.영업상태명 != '취소/말소/만료/정지/중지']
part1.head()
```

```python
# 컬럼별로 중복데이터 여부 확인

part1['영업상태명'].value_counts()
```

```python
# 컬럼 삭제

part1 = part1.drop(['영업상태명'], axis=1)
part1.head()
```

```python
# 컬럼 추가

part1['행정시'] = np.nan
part1.head()
```

```python
# 컬럼명 순서변경

part1 = part1[['사업장명', '업종', '행정시', '도로명주소', '위도', '경도']]
part1.head()
```

```python
# 누락 데이터 확인

part1.count()
```

```python
# 누락 데이터 확인

part1.isnull().sum()
```

```python
# csv 파일 저장

part1.to_csv('제주특별자치도_대규모점포.csv', encoding='cp949')
```
