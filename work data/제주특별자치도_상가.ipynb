```python
import pandas as pd
import numpy as np
  
part2 = pd.read_csv('/content/drive/MyDrive/국민대학원/졸업프로젝트/데이터/1순위/음식/제주특별자치도상가(상권)정보.csv', encoding='cp949')
```

```python
# part2 = 제주특별자치도상가(상권)정보

print(part2.shape)
print(list(part2.columns))
part2.head()
```

```python
# 컬럼별로 중복데이터 여부 확인

part2['상권업종대분류명'].value_counts()
```

```python
# '영업상태명' 컬럼에 있는 '음식'이 들어있는 행 추출

part2 = part2[part2.상권업종대분류명 == '음식']
part2.head()
```

```python
# 컬럼 삭제

part2 = part2.drop(['상권업종대분류명', '상권업종소분류명', '시도명', '행정동명', '법정동명', '행정동코드', '법정동코드', '지번주소', '신우편번호'], axis=1)
part2.head()
```

```python
# 컬럼명 변경

part2.rename(columns = {'상호명':'업소명', '상권업종중분류명':'업태', '표준산업분류명':'업종', 
                        '시군구명':'행정시'},inplace=True)
part2.head()
```

```python
# 컬럼 추가

part2['전화번호'] = np.nan
part2.head()
```

```python
# 컬럼명 순서변경

part2 = part2[['업소명', '업종', '업태', '행정시', '도로명주소', '전화번호', '위도', '경도']]
part2.head()
```

```python
# 누락 데이터 확인

part2.count()
```

```python
# 누락 데이터 확인

part2.isnull().sum()
```

```python
# 완전히 동일한 중복데이터 여부 확인

part2.duplicated(['업소명', '도로명주소']).sum()
```

```python
# '업소명', '소재지' 컬럼인 중복된 행 삭제

part2.drop_duplicates(['업소명', '도로명주소'])
```

```python
# csv 파일 저장

part2.to_csv('제주특별자치도_상가.csv', encoding='cp949')
```
